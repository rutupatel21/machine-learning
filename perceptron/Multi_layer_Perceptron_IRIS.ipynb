```python
# Importing dependencies.
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

```python
# read dataset
data = pd.read_csv('D:/academics/MLpractice/iris.csv')

data["species"].value_counts()

# Statistics for each feature
# No missing values because the total number of instances for each feature is 150(i.e. Count)
data.describe()
```

```python
# Splitting the dataset into the Training set and Test set

from sklearn.model_selection import train_test_split
X = data[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y = data['species']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
# Import the model
from sklearn.neural_network import MLPClassifier

# Outputs:
MLPClassifier (activation='relu', alpha=0.0001, batch_size='auto', beta_1=0.9, 
beta_2=0.999, early_stopping=False, epsilon=1e-08,       
hidden_layer_sizes=10, learning_rate='constant',      
learning_rate_init=0.01, max_iter=500, momentum=0.9,       
nesterovs_momentum=True, power_t=0.5, random_state=None,       
shuffle=True, solver='sgd', tol=0.0001, validation_fraction=0.1,       
verbose=False, warm_start=False)


# Initializing the multilayer perceptron
#mlp = MLPClassifier(hidden_layer_sizes(10),solver='sgd',learning_rate_init= 0.01, max_iter=500)
mlp = MLPClassifier(solver='sgd',learning_rate_init= 0.01, max_iter=500)

# Train the model
mlp.fit(X_train, y_train)

# Predict using the multi-layer perceptron classifier
y = mlp.predict(X_test)
```

```python
# Data visualisation
# Score takes a feature matrix X_test and the expected target values y_test. 
# Predictions for X_test are compared with y_test
print ("accuracy is : ")
print (mlp.score(X_test,y_test, sample_weight=None))
print ()
from sklearn.metrics import confusion_matrix, classification_report, precision_score
from sklearn.metrics import classification_report 

actual = y_test 
predicted = y

results = confusion_matrix(actual, predicted) 
print ('Confusion Matrix :')
print(results)
print ()

print ('Classification Report : ')
print (classification_report(actual, predicted))

```
