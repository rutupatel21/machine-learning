```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn import datasets
from sklearn import svm
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
#Load Iris dataset with sklearn tool (sci-kit learn)
iris = datasets.load_iris()
print("Data Shape and Target Shape\n",iris.data.shape,iris.target.shape)
print("\nSplitting data")
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3, random_state=0)
print ("X training Shape:",X_train.shape,"\tY training shape:",y_train.shape)
print ("X Testing Shape:",X_test.shape, "\tY testing Shape:",y_test.shape)
print("\nSVM Classification on IRIS Dataset")
clf = svm.SVC(kernel='linear', C=1).fit(X_train, y_train)
print("The accuracy of Test Data: ",clf.score(X_test, y_test)*100)
print("The accuracy of Train Data: ",clf.score(X_train, y_train)*100)
pred=clf.predict(X_test)
print("\nClasification Report\n",classification_report(y_test, pred))
print("\nConfusion Report\n",confusion_matrix(y_test,pred))
x_index = 0
y_index = 1
# this formatter will label the colorbar with the correct target names
formatter = plt.FuncFormatter(lambda i, *args: iris.target_names[int(i)])
plt.figure(figsize=(8, 6))
plt.scatter(iris.data[:, x_index], iris.data[:, y_index], c=iris.target)
plt.colorbar(ticks=[0, 1, 2], format=formatter)
plt.xlabel(iris.feature_names[x_index])
plt.ylabel(iris.feature_names[y_index])
plt.tight_layout()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
plt.clf()
fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig, elev=-150, azim=110)
X_reduced = PCA(n_components=3).fit_transform(iris.data)
ax.scatter(X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2], c=iris.target,cmap=plt.cm.Set1, edgecolor='k', s=40)
ax.set_title("First three PCA directions")
ax.set_xlabel("1st eigenvector")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd eigenvector")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd eigenvector")
ax.w_zaxis.set_ticklabels([])
```
